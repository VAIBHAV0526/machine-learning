```python
import seaborn as sns
import numpy as np
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
sns.barplot(x="sex",y="total_bill",data=tips, estimator=np.std )
```

```python
sns.countplot(x="sex",data=tips,hue="smoker",color="red")
```

```python
ax = sns.boxplot(x="tip",y="total_bill",data=tips,hue="sex",color="red")
```

```python
#distrubtion'
sns.displot(tips["total_bill"])
```

```python
import matplotlib.pyplot as plt

sns.kdeplot(tips["total_bill"]) # seams likere normal distrubution'
mean=np.mean(tips["total_bill"])
print(mean)
std=np.std(tips["total_bill"])
print(std)
```

```python
#pair plot
sns.jointplot(x="size",y="smoker",data=tips,kind="reg")
```

```python
sns.pairplot(tips)
```
