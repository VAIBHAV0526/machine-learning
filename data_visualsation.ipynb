```python
import matplotlib.pyplot as plt
import numpy as np
import  math as math
plt.style.use("seaborn")
```

```python
#creating a array 
#line plot
x=np.arange(10) # created a array containg  number from (1,10)
y=2*x+1 # m=2, c=1
plt.plot(x,y,color="red") # to draw the plot
plt.show() # to represnt the curve
```

```python
# lets create a sin and cos curve
time = np.arange(0, 10, 0.1)
print(time)  
y_3=0*time+0;
# valur range from 0,10 in array by incremet of 0.1
amplitude= np.sin(time)
plt.plot(time,y_3)
plt.plot(time,amplitude)
amplitude_2=np.cos(time)

plt.plot(time,amplitude_2)

plt.show()
```
